### Ingredients Analysis

```python
ingredient = (
                df['pizza_ingredients']
                .str.split(',')
                .explode()
                .str.strip()
                .value_counts()
                .reset_index()
                .rename(columns = {'index' : 'Counts' , 'pizza_ingredients': 'Ingredients'})
    
             )

print(ingredient.head(15))
```

Insight:

Garlic, Tomatoes, Red Onions, and Red Peppers are the most frequently used ingredients across the menu. These appear consistently in top-selling pizzas, indicating that customer preference heavily leans toward savory and aromatically rich flavors. High-frequency ingredients should be prioritized in procurement planning to avoid stockouts, while low-frequency items (e.g., specialty cheeses) could be ordered in smaller batches to reduce waste

### Daily Trend - Total Orders

```python
sns.set_theme(style="whitegrid")

weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['day_name'] = pd.Categorical(df['order_date'].dt.day_name(), 
                                categories=weekday_order, 
                                ordered=True)

orders_by_day = df.groupby('day_name', observed=False)['order_id'].nunique()

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=orders_by_day.index,
    y=orders_by_day.values,
    palette="Greens_r",
    edgecolor=None
)

plt.title("Total Orders by Day of Week", fontsize=18, weight='bold')
plt.xlabel("Day of Week", fontsize=13)
plt.ylabel("Number of Orders", fontsize=13)
plt.xticks(rotation=45, fontsize=11)
plt.yticks(fontsize=11)

ax.grid(axis='y', linestyle='--', alpha=0.4)

for i, val in enumerate(orders_by_day.values):
    ax.text(i, val + 30,
            f"{val:,}",
            ha='center', va='bottom', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()
```

Insight:

Friday and Saturday show the highest number of orders, confirming strong weekend demand driven by leisure dining behavior. Midweek days (particularly Tuesday and Wednesday) show the lowest activity. This pattern suggests staffing should be scaled up for weekends, while targeted weekday promotions could help boost midweek sales.

### Daily Trend - Total Revenue

```python
sns.set_theme(style="whitegrid")

weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['day_name'] = pd.Categorical(df['order_date'].dt.day_name(), categories=weekday_order, ordered=True)

revenue_by_day = df.groupby('day_name', observed=False)['total_price'].sum()

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=revenue_by_day.index,
    y=revenue_by_day.values,
    palette="Blues_r",
    edgecolor=None
)

plt.title("Total Revenue by Day of Week", fontsize=18, weight='bold')
plt.xlabel("Day of Week", fontsize=13)
plt.ylabel("Revenue ($)", fontsize=13)
plt.xticks(rotation=45, fontsize=11)
plt.yticks(fontsize=11)

ax.grid(axis='y', linestyle='--', alpha=0.4)

ax.get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, _: f'${x:,.0f}')
)

for i, val in enumerate(revenue_by_day.values):
    ax.text(i, val + 1500, 
            f"${val:,.0f}", 
            ha='center', va='bottom', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()
```

Insight:

Revenue closely mirrors order volume, with Fridays and Saturdays generating the highest sales. This reinforces the importance of weekend operational readiness. Additionally, revenue dips midweek highlight opportunities for bundle deals, discounts, or loyalty incentives to drive transactions on slower days.

### Hourly Trend - Total Orders

```python
sns.set_theme(style="whitegrid")

df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S')
df['order_hour'] = df['order_time'].dt.hour

orders_by_hour = df.groupby('order_hour', observed=False)['order_id'].nunique()

plt.figure(figsize=(12,6))
ax = sns.barplot(
    x=orders_by_hour.index,
    y=orders_by_hour.values,
    palette="Reds_r",
    edgecolor=None
)

plt.title("Hourly Orders Trend", fontsize=18, weight='bold')
plt.xlabel("Hour of Day", fontsize=13)
plt.ylabel("Number of Orders", fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

ax.grid(axis='y', linestyle='--', alpha=0.4)

for i, val in enumerate(orders_by_hour.values):
    ax.text(i, val + 20, f"{val:,}", 
            ha='center', va='bottom', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()
```

Insight:

Ordering behavior shows two strong peaks: lunch (12 PM1 PM) and dinner (6 PM7 PM). Minimal orders occur early morning and late night. This indicates highly predictable demand cycles, enabling clear staffing and prep-time scheduling. Prepping ingredients earlier in the morning and mid-afternoon would optimize workflow and speed of service.

### Monthly Trend - Total Orders

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst = True)

df['month_name'] = df['order_date'].dt.month_name()

month_order = ["January", "February", "March", "April", "May", "June", "July", 
               "August", "September", "October", "November", "December"]

df['month_name'] = pd.Categorical(df['month_name'], categories = month_order, ordered = True)

orders_by_month = df.groupby('month_name', observed = False)['order_id'].nunique()

plt.figure(figsize = (10,5))
plt.fill_between(orders_by_month.index, orders_by_month.values, color = "orange", alpha = 0.6)
plt.plot(orders_by_month.index, orders_by_month.values, color = "black", linewidth = 2, marker = 'o')

plt.title ("Total Orders by Month")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.xticks(rotation = 45)

for i, val in enumerate(orders_by_month):
    plt.text(i, val + 20, str(val), ha = 'center', va = 'bottom', fontsize = 9, fontweight = 'bold')
    
plt.tight_layout()
plt.show()
```

Insight:

Orders steadily rise throughout the year, with noticeable peaks in July, August, October, and December. Summer increases may relate to school breaks and social gatherings, while October and December likely reflect holiday-related dining. These insights support seasonal inventory adjustments and marketing campaigns aligned with holiday and summer peaks.

### % Sales per Category

```python
category_sales = df.groupby('pizza_category')['total_price'].sum()

category_pct = category_sales / category_sales.sum() * 100

plt.figure(figsize = (7,7))
colors = plt.get_cmap('tab20').colors

plt.pie(category_pct, labels = category_pct.index, autopct = '%1.1f%%', startangle = 90, colors = colors, 
       wedgeprops = {'edgecolor' : 'black' , 'width':0.5 })
                    
plt.title("Perecentage of Sales by Pizza Category")
plt.show()
```

Insight:

The Classic category drives the largest share of revenue, followed by Veggie and Supreme. This shows customers strongly prefer familiar, traditional flavors. Chicken pizzas contribute the least, suggesting an opportunity to revamp or reposition Chicken offerings.

### % Sales by Pizza Size and  Category

```python
sales_pivot = df.pivot_table(
    values='total_price',
    index='pizza_category',
    columns='pizza_size',
    aggfunc='sum',
    fill_value=0
)

sales_pct = sales_pivot / sales_pivot.sum().sum() * 100

plt.figure(figsize=(10,6))
sns.heatmap(
    sales_pct,
    cmap="YlOrRd",
    annot=True,
    fmt=".1f",
    linewidths=0.5,
    cbar_kws={'label': 'Revenue ($)'}
)
    
plt.title("Revenue Heatmap by Category and Size", fontsize=16, weight='bold')
plt.xlabel("Pizza Size")
plt.ylabel("Pizza Category")

plt.show()
```

Insight:

Large pizzas consistently generate the highest revenue across all categories, followed by Medium sizes. This indicates that customers perceive larger sizes as offering better value. Small sizes underperform across all categories, suggesting potential re-pricing, repositioning, or removal.

# Key Insights  

## Sales & Demand  
- Total revenue: **$817.8K**, with **49.5K pizzas sold** across **21.3K orders**.  
- Clear demand peaks at **121 PM** (lunch) and **67 PM** (dinner).  
- **Friday & Saturday** generate the highest orders.  
- Seasonal spikes in **July, August, October, and December**.

## Menu Performance  
- **Classic** category leads in both revenue and quantity.  
- **Large pizzas** are the highest revenue drivers in every category.  
- Top performers: **Thai Chicken**, **Classic Deluxe**, **Hawaiian**, **Five Cheese**.  
- Bottom performers: **Brie Carre**, **Mediterranean**, **Soppressata**, **Calabrese**.

## Inventory & Ingredients  
- Most-used ingredients: **Garlic**, **Tomatoes**, **Red Onions**, **Red Peppers**.  
- These overlap strongly with top-selling pizzas  critical for inventory planning.

---

# Recommendations  

## Menu & Product  
- Promote best-selling pizzas (combos, most popular, upsells).  
- Redesign or retire weak-performing specialty pizzas.  
- Expand classic & chicken-based offerings where demand is strong.

## Marketing  
- Run **weekday promotions** (MonWed) to boost off-peak sales.  
- Push premium bundles during peak lunch/dinner hours.  
- Feature top sellers prominently on delivery platforms.

## Operational Efficiency  
- Increase staffing during **121 PM** and **67 PM** peaks + weekends.  
- Pre-prep ingredients during low-demand hours.  
- Streamline inventory for high-frequency ingredients.

## Inventory Optimization  
- Maintain buffer stock of Garlic, Tomatoes, Onions, Peppers.  
- Reduce ordering for niche ingredients tied to low performers.
